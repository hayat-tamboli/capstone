```python
fields = ['Year','Disease','Country']
df_slice = df[fields]
df_summary = df_slice.groupby(fields).count()
df_summary.head(50)
```

```python
df_slice.columns
a = df_slice.apply(lambda x: x.Disease + " " + x.Country,axis=1)
a.value_counts()
```

```python
b = pd.concat([a,df_slice.Year],axis=1)
b.columns = ["Disease_Country","Year"]
b
```

```python
grouped_by_year = b.groupby('Year')
concat_disease = grouped_by_year.sum()
```

```python
transactions = b.groupby(['Year'])['Disease_Country'].apply(list).values.tolist()
transactions
```
